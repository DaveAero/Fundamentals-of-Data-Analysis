Create box plots on a single set of axes for all five experiments in the morley data set.

<h2 style="color: rgb(0, 91, 94);">Answer 1</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

```python
# Loading data from CSV file
morley = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/morley.csv', skiprows = 1, names=['Index', 'Expt', 'Run', 'Speed'])

# Looking at database
print(morley)
```

```python
# Seperating out data into Expt 1 - 5
expt1 = morley[morley['Expt'] == 1]
expt2 = morley[morley['Expt'] == 2]
expt3 = morley[morley['Expt'] == 3]
expt4 = morley[morley['Expt'] == 4]
expt5 = morley[morley['Expt'] == 5]
```

```python
# Create a box plot.

# Adding an Axes for the lables
ax = plt.axes()
# Adding all 5 box plots
plt.boxplot(np.array([expt1['Speed'], expt2['Speed'], expt3['Speed'], expt4['Speed'], expt5['Speed']]).T);
# Adding custom lables to the X axes
ax.set_xticklabels(['Expt 1', 'Expt 2', 'Expt 3', 'Expt 4', 'Expt 5'])
# Adding a global lable to the Y axes
plt.ylabel("Speed (m/s)")

plt.show()
```

```python
# Loading data from CSV file
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'FlowerClass'])

# looking at database
print(iris)
```

```python
# Flower groups
setosa = iris[iris['FlowerClass'] == 'Iris-setosa']
versicolor = iris[iris['FlowerClass'] == 'Iris-versicolor']
virginica = iris[iris['FlowerClass'] == 'Iris-virginica']
```

```python
# Using subplots for the 4 variables
plt.subplot(2, 2, 1)
# Putting title on the first plot
plt.title('1. Iris-setosa, 2. Iris-versicolor, 3. Iris-virginica', loc = 'left')
# Adding boxplots
plt.boxplot(np.array([setosa['SepalLength'], versicolor['SepalLength'], virginica['SepalLength']]).T);
# Adding lable to the Y axes
plt.ylabel("Sepal Length (cm)")

plt.subplot(2, 2, 2)
plt.boxplot(np.array([setosa['SepalWidth'], versicolor['SepalWidth'], virginica['SepalWidth']]).T);
plt.ylabel("Sepal Width (cm)")

plt.subplot(2, 2, 3)
plt.boxplot(np.array([setosa['PetalLength'], versicolor['PetalLength'], virginica['PetalLength']]).T);
plt.ylabel("Petal Length (cm)")

plt.subplot(2, 2, 4)
plt.boxplot(np.array([setosa['PetalWidth'], versicolor['PetalWidth'], virginica['PetalWidth']]).T);
plt.ylabel("Petal Width (cm)")

# using tight layout to avoid plots overlapping
plt.tight_layout()

plt.show()
```

```python
###### Taken from notebook ###### 

# Create a series of x vectors.
xs = np.array([np.linspace(i, i + 2.0, 20) for i in range(10)])
```

```python
# Editied this line, 
# Replaced the "-1.0 * xs[i]" with "8.0 + xs[i]" So the plot will start with roughly the same Max y value as the origional plot
# Replaced the "+ 2 * i" with "- 2 * i" so the plot will decrease at the same rate the origional plot increased

# Create a corresponding series of y vectors.
ys = np.array([8.0 + xs[i] - 2 * i for i in range(len(xs))])
```

```python
###### Taken from notebook ###### 

# Add some noise.
ys = ys + np.random.normal(0.0, 1.0, ys.shape)

# Create global list of x and y values.
x = xs.flatten()
y = ys.flatten()

# Plot.
plt.plot(x, y, '.');

# Best fit line slope and y intercept.
m, c = np.polyfit(x, y, 1)

# Create a steadily increasing x range for plotting.
xbestfit = np.linspace(x.min(), x.max(), 1000)
# Create corresponding y values form best fit line.
ybestfit = m * xbestfit + c

# Plot points.
plt.plot(x, y, '.')

# Plot line.
plt.plot(xbestfit, ybestfit);

# Plot the same points, but now with a different colour for each range.
for i in range(len(xs)):
    plt.plot(xs[i], ys[i], '.')
    
# Same plot but with best fit lines for each category.
for i in range(len(xs)):
    # Plot this category.
    plt.plot(xs[i], ys[i], '.')
    # Fit line to category.
    m, c = np.polyfit(xs[i], ys[i], 1)
    # Create steadily increasing x range for plot.
    xbestfit = np.linspace(xs[i].min(), xs[i].max(), 50)
    # Corresponding y values based on best fit line.
    ybestfit = m * xbestfit + c
    # Plot best fit line for category.
    plt.plot(xbestfit, ybestfit, color='lightgrey');
```
